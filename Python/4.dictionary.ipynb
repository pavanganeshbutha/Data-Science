**A Python dictionary is a data structure that stores key/value pairs. Each element of a Python dictionary is a key/value pair. The keys are like an index operator. In a list, the indexes are integers. In a dictionary, the key must be a hashable object. A dictionary cannot contain duplicate keys. Each key maps to one value. A key and its corresponding value form an item (or entry) stored in a dictionary. The data structure is a called a “dictionary” because it resembles a word dictionary, where the words are the keys and the words’ definitions are the values. A dictionary is also known as a map, which maps each key to a value. It enables fast retrieval, deletion, and updating of the value by using the key.**

- Dictionaries are a collection in python which consists of key-value pairs
- Dictionaries are unordered, indexable and mutable
- Dictionaries are defined using `{}`
- Dictionaries don't allow duplicate keys

# 1. Creating a Dictionary
- You can create a dictionary by enclosing the items inside a pair of curly braces ({}). Each item consists of a key, followed by a colon, followed by a value. The items are separated by commas.

**You can create an empty dictionary by using the following syntax:**

```python
dictionary1={"11-11-11":"Alex","11-11-12":"John"}
dictionary1
```

```python
dictionary2={1:"Alex",2:"John"}
dictionary2
```

```python
dictionary3={1:23,2:"John"}
```

```python
dictionary3
```

**Note:Python uses curly braces for sets and dictionaries. The syntax {} denotes an empty dictionary.To create an empty set, use set()**

```python
students={} # Empty dictionary
students
```

### get()
- To prevent a KeyError exception, we can use the get() function on a dictionary. Then, if a value is not found for a given key, it returns nothing (and otherwise returns the value associated with that key).

```python
dictionary3.get(3)
```

```python
dictionary3.get(333)
```
